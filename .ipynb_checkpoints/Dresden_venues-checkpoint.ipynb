# Dresden: where to go?

## Problem description

In this notebook we are going to consider the following problem: let's say we have a tourist who comes to Dresden (Germany) for a short period of time. As any good tourist, he or she wishes to see the best landmarks and visit the best venues. However, there are so many places to go in Dresden! How to decide where to go? Given the fact that our tourist in Dresden only for so long, time is of the essence and he has to choose. <br>
 <br>
 We are going to help our tourist. Using the Foursquare app, we shall obtain some data about Dresden venues (which data to be discussed later). Based on this data, we will divide all venues into three clusters - each of them containing venues which are likely to be similarly appreciated by the tourist. We could call this a recommender system. <br> <br>
So, the steps to be done are the following:
-  collect data about Dresden venues
-  visualize this data in order to have some understanding of it
-  use the data to cluster the venues (KMeans-algorithm) 
-  based on the results of clustering, recommend our tourist where to go first and where - if there is time

This could be interesting for a tourist guide or just anyone who wishes to come do Dresden for a visit - since it is pretty easy to use and interpret.

## Data <a name="dat"></a>

### Which Data? <a name="dat1"></a>

For our purposes we shall use the Foursquare app. We will consider venues no further than 5km from the center of Dresden. <br> 
We need to bear in mind that we need data that reflects people's appreciation of a given venue. Using Foursquare app, we can obtain the following features for each venue:

-  __Rating__: this is one of the most reliable ways to determine whether the venue is good or bad
-  __Number of "likes"__: also reflects pretty well appreciation for the place
-  __Number of comments__: if someone felt that the place is worth of praise (or a whack for that matter), we should consider this
-  __Distance__: our tourist does not have a lot of time. The distance from the center should be factored in our analysis
-  __Difference between the number of "likes" and comments__: if there are more comments than likes, it could indicate that the venue is not so good. Indeed, people who are happy enough with the venue to leave a comment usually leave a like, don't they? Thus, there should be more likes than comments

All of this will be obtained using Foursquare premium calls. By the way, a free developer's account allows only 50 premium calls per day, so I had to collect information for several days. I recorded it in a CSV file "keeping.csv" which shall be submitted in GitHub along with this notebook.

### How to use Data? <a name="dat2"></a>

This is pretty straightforward actually. Once we gather and prepare data, it shall be used to cluster the venues with K-Means algorithm. A weighted version will be used - we want to be able to accord different levels of importance to our features (may be, our tourist does not trust the comments online, or trusts a lot to ratings, or has too little time and thus the distance is of utmost importance - who knows?..)

## Methodology <a name='meth'></a>

The data has been analyzed as follows:
-  Upon collection, the cleanness was verified. Luckily, all values in all rows proved to be valid (no missing values or NaNs or whatnot)
-  Data correlation was studied (Pearson correlation). The results made me drop one of the features I wanted to use due to it's high correlation with another one. 
-  Data was visualized: the objective was to get an idea about the distribution of values of each feature; see if there are outliers (and how many) and if they vary enough to expect informative clustering. Box plots, histograms and probability density functions were used for these purposes.
-  Finally, K-means algorithm was applied in order to divide venues into three clusters. Each cluster contains venues that should have similar appeal to the user. K-means was chosen over other clustering algorithms despite the presence of some outliers (as per Visualization section) because outliers are not a serious issue for this particular problem (as opposed to anomaly detection where we would have to use DBSCAN). 
-  Before K-means was applied, the data was weighted. This is supposed to provide some flexibility to individual customer needs (for different people, one factors may be of greater importance than the others)

Let's begin.

## Collecting the data <a name="coll"></a>

```python
#importing libraries
import pandas as  pd
import numpy as np
import requests
import folium
from pandas.io.json import json_normalize
```

```python
#Foursquare account credentials and Dresden coordinates

CLIENT_ID = 'id' 
CLIENT_SECRET = 'secret' 
VERSION = '20180605'
LIMIT=500
radius=5000

lat= 51.0504
lng= 13.7373
```

```python
 url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
```

```python
#Getting general info about the venues
venues = requests.get(url).json()['response']['groups'][0]['items']
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
nearby_venues = json_normalize(venues)

filtered_columns = ['venue.name','venue.categories','venue.location.lat','venue.location.lng','venue.id','venue.location.distance']
nearby_venues = nearby_venues.loc[:,filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
nearby_venues.info()
```

```python
venues_grouped = nearby_venues[['categories','name']].groupby('categories',as_index=False).count()
venues_grouped.rename(index=str, columns={'name':'count'},inplace=True)
venues_grouped.sort_values(by='count', ascending=False,inplace=True)
venues_grouped
```

We wish to keep only venues that are of some attraction to tourists. Thus things like fitness, soccer or supermarket are of not interest to us. Let's filter them out.

```python
to_filter_out = ['IT Services, Gym / Fitness Center','Electronics Store','Soccer Stadium',
                 'Hotel','Supermarket', 'Hotel Bar','Field']

for elem in to_filter_out:
    nearby_venues = nearby_venues.loc[nearby_venues['categories'] != elem]
    
nearby_venues = nearby_venues[['name','categories','lat','lng','id','distance']].reset_index()
```

```python
dresden_map=folium.Map(location=[lat,lng],zoom_start=12.6)

for lat, lng, label in zip(nearby_venues['lat'], nearby_venues['lng'], nearby_venues['name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        ).add_to(dresden_map)
    
dresden_map
```

Below is the code used to recover data about each specific venus (premium Foursquare calls). The results have been recorded in "keeping.csv" and now can be used without making any calls to outside apps.

```python
# likes = list()
# ratings = list()
# tipsNum = list()

# ids = nearby_venues['id'].tolist()
# print(len(ids))
# cnt = 0
# for id_ in ids[46:len(ids)]:
#     print('current id: ' + str(id_))
#     print(cnt)
#     cnt+=1
#     url_specific = 'https://api.foursquare.com/v2/venues/{}?&client_id={}&client_secret={}&v={}'.format(
#             id_,
#             CLIENT_ID, 
#             CLIENT_SECRET, 
#             VERSION
#             )
#     venue = requests.get(url_specific).json()
#     likes.append(venue['response']['venue']['likes']['count'])
#     ratings.append(venue['response']['venue']['rating'])
#     tipsNum.append(venue['response']['venue']['stats']['tipCount'])

venue_info = pd.read_csv('keeping.csv')
venue_info['comment_like_diff'] = venue_info['tipcount']-venue_info['likes']
venue_info.head()
```

## Exploratory Data Analysis <a name="eda"></a>

### Data cleanness and correlation <a name='corr'></a>
We are going to cluster the venues, and we would like to use the five parameters we gathered above - their rating, amount of tips, amount of likes, distance from venues to town center and difference between the numbers of tips and likes. But before we do that, let's see if these features are actually informative. Remember, our goal is to help a tourist decide which venues are to be visited first, second and so on.

```python
feature_df = venues_global[['distance','rating','likes','tipcount','comment_like_diff']]
feature_df.info()
```

So, we have no missing or corrupted data. That is good. <br>
Let's see now if some of these features are strongly correlated (and thus redundant)

```python
feature_df.corr()
```

Okay, so, evidently, comment_like_diff and likes are very much correlated. The idea behind comment_like_diff was that usually when there are more comments than likes, it means that the venue is not that great - angry people live comments with no like, happy people leave either a like or a like+comment (normally). Let's try to change this column: if it is positive, we will put 1, if negative - 0

```python
feature_df['comment_like_diff'].apply(lambda x: 1 if x>0 else 0).value_counts().to_frame()
```

Oh. So, there is always more likes than comments. Well, there goes my idea. This column is not informative (strongly correlated with likes and all negative), we shall delete it.

```python
feature_df = feature_df[['distance','rating','likes','tipcount']]
feature_df.corr()
```

tipcount and likes are also correlated, but I would still like to keep both of them. 

### Visualization <a name="vis"></a>

How are the values of our features distributed? Do they differ enough to provide informative clustering? We will get an idea of it by using box plots, histograms and distribution densities

```python
import matplotlib.pyplot as plt
```

```python
fig1 = plt.figure(1)
for col in feature_df.columns.tolist(): 
    fig1.add_subplot(2,2,feature_df.columns.tolist().index(col)+1)
    feature_df[col].plot(kind='box',figsize=(15,10),color='blue')
    plt.title("Box plot of " + str(col))
    plt.ylabel(str(col) + " value")
```

Some of the places have way more likes and tips than the rest, don't they. K-means algorithm is not impervious to outliers, so I we should take a look at what we are dealing with. 

```python
venues_global.sort_values(['likes'], ascending = False).head()
```

```python
venues_global.sort_values(['tipcount'],ascending=False).head()
```

While we could use some outliers-sensitive algorithm instead of K-means (DBSCAN for example), in this case the outliers problem is not such a big deal. First of all, we are not trying to detect any kind of anomaly, so it is not that important to know which points are outliers and which are not. Secondly, there are just a couple of outliers in each category (as per the box plots), so I doubt that they will do a lot of damage. Let's continue.

```python
#Plotting some histograms 
feature_df.hist(color='red',alpha=0.5,bins=20,figsize=(10,7))
```

```python
#plotting density functions
fig2 = plt.figure(2,)
for col in feature_df.columns.tolist():
    fig2.add_subplot(2,2,feature_df.columns.tolist().index(col)+1)
    feature_df[col].plot.kde(figsize=(10,7))
    plt.title(col)
    plt.ylabel(col+ ' density')
    
```

It seems that "likes" column will not contribute much to the clustering - probabilty density function is very narrow, and on the box plot first three quartiles are very close to each other. So, if we did clustering based on likes alone, we would probably get one big cluster and some very small ones. 
Rating, however, seems interesting. Pretty good feature, I think.


## Clustering <a name='clust'></a>

We shall employ K-means clustering to divide the venues into three categories. Hopefullsy, this would give a good indication as to what places are the best choices for tourists.

```python
from sklearn.cluster import KMeans 
from sklearn.preprocessing import StandardScaler
```

We shall give weights to our features. Some people may trust in likes above everything else, some may favor rating. Others might have too little time, so for them distance is very important. With weights, we could tailor the algorithm to individual needs.

```python
feature_df = feature_df[['distance','rating','likes','tipcount']]

DIST_W = 3
RATE_W = 5
LIKE_W = 1
TIP_W  = 2

weights = np.array([DIST_W, RATE_W, LIKE_W, TIP_W])

#Scaling features
X = feature_df.values[:,:]
clustering_dataset = StandardScaler().fit_transform(X)

#Applying weights
for elem in clustering_dataset:
    for i in range(0,4):
        elem[i] = elem[i]*weights[i]
```

```python
num_clusters = 3
k_means = KMeans(init="k-means++",n_clusters=num_clusters,n_init=12)
k_means.fit(clustering_dataset)
labels=k_means.labels_
print(labels)
```

## Viewing the results <a name='view'></a>
Let's take a look at what we got.

```python
venues_global['cluster_num'] = pd.Series(labels)
venues_global.head()
```

```python
result_map=folium.Map(location=[lat,lng], zoom_start=14)
colors = ['yellow','red','green']

for lat, lng, label, rating, likes,color in zip(venues_global['lat'], venues_global['lng'],
                                                venues_global['name'],venues_global['rating'], venues_global['likes'],
                                                venues_global['cluster_num']):
    label = folium.Popup(str(label)+ ",rating: " + str(rating) + ",likes: " + str(likes), parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=colors[color],
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        ).add_to(result_map)
    
result_map
```

In this particlar case, we judged the venues primaraly by their rating. It seems to have worked :) 

## Results <a name='res'></a>

The K-means did it's job - clusters seem to be logical to the human eye. <br>
Tweaking the weights of each feature gives expected results - "heavier" features start to matter more than others, changing the clusters.<br>
